```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, tight_layout=True)

#折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数
x = np.linspace(0, 10, 200)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y)

#ヒストグラム
x = np.random.normal(30, 5, 3000)
axes[0,2].hist(x, bins=30)

#散布図
np.random.seed(4)
x = 5 + np.random.normal(0, 3, 100)
y = 5 + np.random.normal(0, 3, len(x))
axes[1,0].scatter(x, y)

plt.show()
```
